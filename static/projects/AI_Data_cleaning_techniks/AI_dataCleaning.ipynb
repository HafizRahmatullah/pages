```python


# Get the column names for each DataFrame
columns_df1 = list(df1.columns)
columns_df2 = list(df2.columns)
columns_df3 = list(df3.columns)
columns_df4 = list(df4.columns)

# Function to compare columns index by index
def compare_columns(columns1, columns2):
    differences = []
    for i, (col1, col2) in enumerate(zip(columns1, columns2)):
        if col1 != col2:
            differences.append((i, col1, col2))
    return differences

# Check for differences
differences_df1_df2 = compare_columns(columns_df1, columns_df2)
differences_df1_df3 = compare_columns(columns_df1, columns_df3)
differences_df1_df4 = compare_columns(columns_df1, columns_df4)
differences_df2_df3 = compare_columns(columns_df2, columns_df3)
differences_df2_df4 = compare_columns(columns_df2, columns_df4)
differences_df3_df4 = compare_columns(columns_df3, columns_df4)

# Print the differences
def print_differences(differences, name1, name2):
    if differences:
        print(f"Differences between {name1} and {name2}:")
        for diff in differences:
            print(f"Index {diff[0]}: {diff[1]} != {diff[2]}")
    else:
        print(f"No differences between {name1} and {name2}")

print_differences(differences_df1_df2, 'df1', 'df2')
print_differences(differences_df1_df3, 'df1', 'df3')
print_differences(differences_df1_df4, 'df1', 'df4')
print_differences(differences_df2_df3, 'df2', 'df3')
print_differences(differences_df2_df4, 'df2', 'df4')
print_differences(differences_df3_df4, 'df3', 'df4')
```

```python
d['Average Temperature (2 m)'] =d['Average Temperature (2 m)'].astype(str).str.strip('°C ')
d['Average Temperature (2 m)'] =d['Average Temperature (2 m)'].astype(str).str.replace(' ','')
d['Average Temperature (2 m)'] =d['Average Temperature (2 m)'].astype(str).str.replace(' to ','-')
split_columns = d['Average Temperature (2 m)'].str.split('-', expand=True)
d[['Average Temperature (2 m)_1', 'Average Temperature (2 m)_2']] = split_columns
d['Average Temperature (2 m)_1'].unique()
```

```python
d['Average Precipitation per day'] =d['Average Precipitation per day'].astype(str).str.strip(' mm ')

split_columns = d['Average Precipitation per day'].str.split('-', expand=True)
d[['Average Precipitation per day_1', 'Average Precipitation per day_2']] = split_columns
d['Average Precipitation per day_1'].unique()
```

```python
d['Average Soil Temperature (0 to 6 cm)'] =d['Average Soil Temperature (0 to 6 cm)'].astype(str).str.strip(' °C ')

split_columns = d['Average Soil Temperature (0 to 6 cm)'].str.split('-', expand=True)
d[['Average Soil Temperature (0 to 6 cm)_1', 'Average Soil Temperature (0 to 6 cm)_2']] = split_columns
d['Average Soil Temperature (0 to 6 cm)_1'].unique()
```

```python

d['Average Sunshine Duration'] =d['Average Sunshine Duration'].astype(str).str.replace(' hours','h')
d['Average Sunshine Duration'] =d['Average Sunshine Duration'].astype(str).str.replace(' seconds','s')
split_columns = d['Average Sunshine Duration'].str.split('-', expand=True)
d[['Average Sunshine Duration_1', 'Average Sunshine Duration_2']] = split_columns
d['Average Sunshine Duration'].unique()

```

```python
d['Humidity(%)'] =d['Humidity(%)'].astype(str).str.strip(' % ')

split_columns = d['Humidity(%)'].str.split('-', expand=True)
d[['Humidity(%)_1', 'Humidity(%)_2']] = split_columns
d['Humidity(%)_1'].unique()
```

```python
d['Rainfall Alert Maximum Rain (mm)'] =d['Rainfall Alert Maximum Rain (mm)'].astype(str).str.strip(' mm ')

split_columns = d['Rainfall Alert Maximum Rain (mm)'].str.split('-', expand=True)
d[['Rainfall Alert Maximum Rain (mm)_1', 'Rainfall Alert Maximum Rain (mm)_2']] = split_columns
d['Rainfall Alert Maximum Rain (mm)_1'].unique()
```
